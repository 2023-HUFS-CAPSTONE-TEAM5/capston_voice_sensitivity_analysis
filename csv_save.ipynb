```python
import glob
import json
import csv

file_list = sorted(os.listdir(dataRoot))
dtRoot = os.path.join(dataRoot, file_list[0])
file_dir_list = sorted(os.listdir(dtRoot))
note = [0*len(file_dir_list)]
for j in range(len(file_dir_list)):
    note = glob.glob(os.path.join(os.path.join(dtRoot, file_dir_list[j]), '*.*'))
    for ii in range(len(note)):
        with open(note[ii], 'r') as f:
            json_data = json.load(f)
        if json_data['화자정보']['Gender'] == 'Male':
            f = open("/content/drive/MyDrive/apple_farm/original/label/test_male.csv", 'a', newline='')
            wr = csv.writer(f)
            wr.writerow("01-" +[note[ii].split("/")[-1], "Male"])
            f.close()
        else:
            f = open("/content/drive/MyDrive/apple_farm/original/label/test_female.csv", 'a', newline='')
            wr = csv.writer(f)
            wr.writerow("01-"[note[ii].split("/")[-1], "Female"])
            f.close()
```
