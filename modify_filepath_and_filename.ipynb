```os module``` 사용하여 파일을 이동하지 않고 새로운 파일명을 가진 파일을 original/voice/{감정라벨-감정} 경로에 저장함  
⇒ 일부 데이터가 새로 저장되지 않으며, 어떤 데이터가 누락되었는지 알 수 없는 문제 발생




````python
angry = TS1_emotion_folders[0]
angry_wav_files = glob.glob(angry + '/*/*.wav')

new_path = "/content/drive/MyDrive/apple_farm/original/voice/05_angry/" 

for wav_file in angry_wav_files:
  new_file_name = "05-" + wav_file.split('/')[-1]
  shutil.move(wav_file, new_path + new_file_name)
  print(f'{new_file_name} has been moved to {new_path}')
````
